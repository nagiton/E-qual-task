# 概要
LSTM、BiLSTM、attentionの3つのモデルで英→日のNMTモデルを作成した。

それぞれの性能について比較する。

# 結論
LSTM、BiLSTMは学習がうまく進まず、ほとんど使い物にならないモデルになった。

一方でattentionを使ったモデルはBLEUスコアは平均値で0.2程度で、やや単語に齟齬があったりするが、構文としては正解に近いものが得られた。

# 可視化

```python
import pandas as pd

df_BiLSTM = pd.read_csv('BiLSTM-0202.csv')
df_LSTM = pd.read_csv('LSTM-0202.csv')
df_attention = pd.read_csv('default-0201.csv')
```

```python
% matplotlib inline
```

## BLEUスコア

```python
df_BLEU = pd.DataFrame()
df_BLEU['Epoch'] = df_attention['Epoch']
df_BLEU['attention'] = df_attention['BLEU']
df_BLEU['LSTM'] = df_LSTM['BLEU']
df_BLEU['BiLSTM'] = df_BiLSTM['BLEU']
```

```python
df_BLEU.plot('Epoch')
```

実際の出力もそうだが、LSTMとBiLSTMはほとんど役に立たない一方、attentionはまずまず

# Loss(cross entropy loss)

```python
df_Loss = pd.DataFrame()
df_Loss['Epoch'] = df_attention['Epoch']
df_Loss['attention'] = df_attention['Loss']
df_Loss['LSTM'] = df_LSTM['Loss']
df_Loss['BiLSTM'] = df_BiLSTM['Loss']
```

```python
df_Loss.plot('Epoch')
```

LSTMとBiLSTMにはほとんど差はない。attentionは効率的に学習できている。

# 考察

LSTMもBiLSTMも、デコーダーが隠れ状態を1つしか参照できず、全体を復元するのが厳しかったのではなかろうか。
